The session below was used to determine which data to use as feature. In conclusion, the 25m and 50m data is very similar and there is no need to build two model from 25m and 50m to predict coastline. For the 6 parameters (['MWD','PWP','SWH','TWD','WiD','WiS']), there is no significant correlation(except wind speed and SWH). There is no harm to use all parameters as features.

```python
#shall i use both 25 and 50? explore the similarity between 25m and 50m data
swh25 = np.transpose(arr25, (0, 2, 1, 3))
shape = swh25.shape
swh25f = np.reshape(swh25, (shape[0]*shape[1], shape[2],shape[3]))
mean25 = np.mean(swh25f,axis=0)

swh50 = swh25 = np.transpose(arr50, (0, 2, 1, 3))
swh50f = np.reshape(swh50, (shape[0]*shape[1], shape[2],shape[3]))
mean50 = np.mean(swh50f,axis=0)

x = np.arange(shape[2])
fig = plt.figure(figsize=(12, 8))


# Generate six plots for the last dimension
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.plot(x, mean25[:, i], label='25m')
    plt.plot(x, mean50[:, i], label='50m')
    plt.xlabel('time step')
    plt.ylabel('Values')
    plt.title(f'Plot for {cols[i]}'.format(i))
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation coefficient between columns for 25m

# Calculate the correlation coefficient between columns
corr_matrix = np.corrcoef(flattened_data25, rowvar=False)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the correlation coefficient matrix as a heatmap
heatmap = ax.imshow(corr_matrix, cmap='coolwarm')

# Set the tick labels and positions for the x-axis and y-axis
ax.set_xticks(np.arange(len(cols)))
ax.set_yticks(np.arange(len(cols)))
ax.set_xticklabels(cols, rotation=90)
ax.set_yticklabels(cols)

# Add a colorbar
cbar = plt.colorbar(heatmap)

# Set the title and labels
ax.set_title('Correlation Coefficient Matrix')
ax.set_xlabel('Columns')
ax.set_ylabel('Columns')

# Show the plot
plt.tight_layout()
plt.show()
```
